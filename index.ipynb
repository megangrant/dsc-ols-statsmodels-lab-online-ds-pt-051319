## Step 4: Run a simple regression in Statsmodels with TV as a predictor

```python
# import libraries
import statsmodels.api as sm
import statsmodels.formula.api as smf

# build the formula 
f = 'sales ~ TV'

# create a fitted model in one line
model = smf.ols(formula = f, data = df).fit()
```

## Step 5: Get Regression Diagnostics Summary

```python
model.summary()
```

Note here that the coefficients represent associations, not causations

## Step 6:  Draw a prediction line with data points on a scatter plot for X (TV) and Y (Sales)

Hint: You can use the `model.predict()` function to predict the start and end point of of regression line for the minimum and maximum values in the 'TV' variable. 

```python

# create a DataFrame with the minimum and maximum values of TV
df2 = pd.DataFrame({'TV': [df.TV.min(), df.TV.max()]})

# make predictions for those x values and store them
df3 = model.predict(df2)
# first, plot the observed data and the least squares line
df.plot(kind = 'scatter', x='TV', y='sales')
plt.plot(df2, df3, color ='red')
plt.show()
```

## Step 7: Visualize the error term for variance and heteroscedasticity

```python
fig = plt.figure(figsize = (15,8))
fig = sm.graphics.plot_regress_exog(model, "TV", fig=fig)
plt.show()
```

```python
# Record Your observations on heteroscedasticity
# The data looks heteroscedastic to me, as the residuals look to increase as TV increases.
# This would therefore mean that the model has failed the linearity assumptions.
```

## Step 8: Check the normality assumptions by creating a QQ-plot

```python
# Code for QQ-plot here
import scipy.stats as stats
residuals = model.resid
fig = sm.graphics.qqplot(residuals, dist = stats.norm, line='45', fit=True)
fig.show()
```

```python
# Record Your observations on the normality assumption
# The data fits the qq plot quite linearly, indicating normal distibution of the data.
# Based on this plot I would pass the normality assumption.
```

## Step 9: Repeat the above for radio and record your observations

```python
# code for model, prediction line plot, heteroscedasticity check and QQ normality check here

#MEG: Creating the model for radio
f2 = 'sales ~ radio'
model2 = smf.ols(formula = f2, data = df).fit()

#MEG: Plot the model vs the underlying data on a graph
df4 = pd.DataFrame({'radio': [df.radio.min(), df.radio.max()]})
df5 = model2.predict(df4)
df.plot(kind = 'scatter', x='radio', y='sales')
plt.plot(df4, df5, color ='red')
plt.show()

#MEG: Visualising the error for radio
fig3 = plt.figure(figsize = (15,8))
fig3 = sm.graphics.plot_regress_exog(model2, "radio", fig=fig3)
plt.show()

#MEG: Code for QQ-plot 
import scipy.stats as stats
residuals = model2.resid
fig2 = sm.graphics.qqplot(residuals, dist = stats.norm, line='45', fit=True)
fig2.show()
```

```python
model2.summary()
```

```python
# Record your observations of good fit here
# MEG:
# From the error visualisation graphs you can immediately see that the data is heteroscedastic, therefore fails the linear assumption.
# Then the qq plot also indicates that the data is not normally distributed, so the normality assumption falls over too.
# The radio data therefore misses 2 of the linearity assumptions and fails.
```
